```python
import requests
import re
import urllib
from bs4 import BeautifulSoup

from playsound import playsound
import os
import subprocess
```

```python
class dictionary:
    def __init__(self):
        ...
        
    def en2kr(self, text):
        text = text.replace("ﬂ", "fl")
        session = requests.Session()
        header = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_5)\
                    AppleWebKit 537.36 (KHTML, like Gecko) Chrome",
                    "Accept":"text/html,application/xhtml+xml,application/xml;\
                    q=0.9,imgwebp,*/*;q=0.8"}

        encText = urllib.parse.quote(text)
        url = "https://dic.daum.net/search.do?q=" + encText
        req = session.get(url, headers=header)
        bs0bj = BeautifulSoup(req.text, "html.parser")

        try:
            box = bs0bj.find("div", class_="cleanword_type kuek_type")
            box = box.find('ul', class_="list_search")
            box = str(box.find_all("span", class_="txt_search"))
            box = re.sub('<.+?>', '', box, 0).strip()
            sound = bs0bj.find("div", class_="wrap_listen")
            sound = sound.find_all("a", class_="btn_voice btn_listen")
            box = box.replace(" ", "").replace("[", "").replace("]", "").replace(",", " ")
            box = box.split(" ")
            
            return box, {"USA" : sound[0]['href'], "UK" : sound[1]['href']}

        except AttributeError:
            return None
    
    def ja2kr(self, text):
        session = requests.Session()
        header = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_5)\
                    AppleWebKit 537.36 (KHTML, like Gecko) Chrome",
                    "Accept":"text/html,application/xhtml+xml,application/xml;\
                    q=0.9,imgwebp,*/*;q=0.8"}

        encText = urllib.parse.quote(text)
        url = "https://dic.daum.net/search.do?q="+ encText + "&dic=jp"
        req = session.get(url, headers=header)
        bs0bj = BeautifulSoup(req.text, "html.parser")
        box = bs0bj.find("div", class_="cleanword_type kujk_type")

        try:
            box = box.find('ul', class_="list_search")
            box = str(box.find_all("span", class_="txt_search"))
            box = re.sub('<.+?>', '', box, 0).strip()
            sound = bs0bj.find("div", class_="search_cleanword")
            sound = sound.find("a", class_="btn_voice btn_listen")['href']
            box = box.replace(" ", "").replace("[", "").replace("]", "").replace(",", " ")
            box = box.split(" ")
            return box, {"JA" : sound}

        except AttributeError:
            return None    
        
    def play_audio_from_url(self, url, fname="1"):

        r = requests.get(url)
        with open(f"./{fname}.mp3",'wb') as f: 
            f.write(r.content)

        try:
            playsound(f"{fname}.mp3")
            os.remove(f"./{fname}.mp3")
        except:
            print("only english")
            
    def print_meaning(self, data):
        for i in range(len(data)):
            print(f"{i+1}. {data[i]}")
```

```python
text="whatever"
dic = dictionary()
data = dic.en2kr(text=text)
usa, uk = data[1].items()
_, url_usa = usa
_, url_uk = uk
dic.print_meaning(data=data[0])
```

```python
dic.play_audio_from_url(url= url_usa)
dic.play_audio_from_url(url= url_uk)
```
